Нам предоставили часть какого-то dataset-а с информацией (в дальнейшем нам дали ещё два, и поставили задачу объединить их). Наша задача состояла в следующем: используя pandas выяснить, что за информация была нам предложена. В ходе работы, мы поняли, что перед нами статистика пассажиров Титаника. В дальнейшем нам предоставили текстовый документ, в котором содержались наименования колонок таблицы, которую мы преобразовали из dataset-ов, посредством библиотек pandas, но колонки были перемешаны и искажены, в придачу были и лишние. Связанная с этим задача состояла в том, чтобы поставить колонки в соответствие столбцам таблицы, привести их в надлежащий вид (перевести на английский язык) и напоследок устранить выбросы в самой таблице. И последней задачей было следующее: на основе данной таблицы выдать рекомендацию бабушке и её внуку: стоит ли им отправляться в путешествие на Титанике.image.png

```python
import pandas
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns #подключили нужные для анализа данных библиотеки
df = pandas.read_csv("2_3.csv")
df.head()
```

```python
 
colums=[ "PassengerId", "Status", "Pclass", "NameSurname", "sex", "age", "SibSp", "Parch", "Ticket", "fare", "cabin", "Embarked"]#Список с названиями колонок
df1=pandas.read_csv("2_1.csv",names=colums)
df2=pandas.read_csv("2_2.csv",names=colums)
df3 = pandas.read_csv("2_3.csv",names=colums)#загрузили 3 датасета , и присвоили им нужные названия колонок для будущего обьединения в один набор данных 
concat_data = pandas.concat([df1, df2,df3], ignore_index=True)#объединили  датасеты в один
s_list=["female","male"]
for i in range(len(concat_data)):#процесс отъсеивания выбросов в колонках возраст и пол
    if concat_data.loc[i,"sex"] not in  s_list:
        concat_data.loc[i,"sex"]=None
    if concat_data.loc[i,'age']>100:
        concat_data.loc[i,'age']=None
```

```python

concat_data.head()#Вывели часть имеющихся данных , что б в будущем понять , с чем будем работать
```

Наше задание состояло в том, чтобы выдать рекомендацию мужчине и его бабушке 85 лет: стоит или не стоит отправляться в путешествие на Титанике исходя из данных, представленных в датасете. За основу наших рекомендаций, мы решили взять статистику смертности. С помощью кода представленного выше мы сделали сортирвоку и выяснили, что в живых из 891-ого человека осталось всего-лишь 342. Погибло 549, что состовляет практически 2/3 от общего числа пассажиров. Статистика удручает, шансы выжить всего около 33%, поэтому уже на данном моменте мы бы крайне не рекомендовали отправляться в плавание на титанике. (1)

```python


survival=concat_data["Status"].value_counts()#cобрали данные о выживших  и погибших посажирах
print(survival)#Вывели полученные данные
sns.factorplot('Status', data=concat_data, kind='count')#на основе датасета построили график выживших и погибших

#(1)
```

Но, давайте попробуем представить, что они решили рискнуть и всё же отправились в плавание на Титанике. Проанализировав остальную информацию, мы составили рекомендоцию как минимзировать риск смертности. Начали мы с класса, которым наши пассажиры решили бы путешествовать. Отсортировав классы, мы узнали их количество: 1-ый класс 216 человек, 2-ой класс 184 человека, 3-ий класс 491 человек. (2)

```python
_class_=concat_data["Pclass"].value_counts()#собрали данные о распределение пассажиров в различных класса
print(_class_)#вывели имеющиеся данные
fg = sns.factorplot('Pclass', data=concat_data, kind='count', aspect=1)#в виде гистограммы вывели расспределение пасажиров  по классам
fg.set_xlabels('Class')

#(2)
```

Затем мы вывели их смертности. Данные следующие: 1-ый класс 80 человек, 2-ой класс 97 человека, 3-ий класс 372 человек. Видно, самый не безопасный класс это самый дешёвый - третий. Погибло более 75-ти процентов пассажиров. В процентном соотношении первый и второй класс не сильно отличаются, в районе 50-ти процентов погибших, но в первом классе всёже чуть меньше, около 45 процентов. Отсюда можно сделать вывод, что если наши клиенты хотят максимизировать шансы на выживание, им придётся потратиться на дорогие билеты. (2)

```python
pc1=0
pc2=0
pc3=0
for  i in range(len(concat_data)):# c помощью метода loc и условного оператора узнали количество погиших в каждом классе 
        if int(concat_data.loc[i,'Pclass'])==1:
            if int(concat_data.loc[i, 'Status']) == 0:
                pc1+=1
        elif int(concat_data.loc[i,'Pclass'])==2:
            if int(concat_data.loc[i, 'Status']) == 0:
                pc2+=1
        elif int(concat_data.loc[i,'Pclass'])==3:
            if int(concat_data.loc[i, 'Status']) == 0:
                pc3+=1
            
print(pc1,pc2,pc3)
#(2)
```

Следующий пункт, который мы рассмотрели, это пол пассажиров. И без данной базы данных мы уже слышали о том, что из-за нехватки спасательных шлюпок, в первую очередь эвокуировали женщин и детей. Следовательно смертность среди мужчин явно должна быть выше. Но слухи слухами, а информацию мы проверили и в датасете. Выяснили, что мужчин было 574, а женщин 312.

Из них мёртвые 466 мужчин и 80 женщин. Результат крайне показательный. (3)

```python
dead_female=0
dead_men=0
for i in range(len(concat_data)):# c помощью перебора узнаем количество погибших мужчин и женщие
    if concat_data.loc[i, 'sex'] == "male":
        if concat_data.loc[i, 'Status'] == 0:
            dead_men += 1
    elif concat_data.loc[i, 'sex'] =="female":
        if concat_data.loc[i, 'Status'] == 0:
            dead_female += 1
print(dead_men,dead_female)
df=[concat_data.loc[i,"sex"] for i in range(len(concat_data)) if concat_data.loc[i,"sex"] != None]
g = sns.factorplot('sex', data=concat_data, hue='Status', kind='count', aspect=1,legend=True)#вывели статистику в виде гистограммы распределения выживаемости среди мужчин и женщин  
g.set_xlabels('Dead')
#Смерти мужчин/женщин (3)
```

Из выживших мужчин, 21 ребёнок до 12 лет. Это наглядно показывает что у мужчин шансы на выживание крайне малы, у женщин ситуация на порядок лучше. (3)

```python
menagenotdead=0

for i in range(len(concat_data)):# перебором собрали данные выживаемости детей 12
    if concat_data.loc[i,"age"]<=12:
        if concat_data.loc[i, 'sex'] == "male":
            if concat_data.loc[i, 'Status'] == 1:
                menagenotdead += 1
print(menagenotdead)
#Живые дети мальчики (3)
```
